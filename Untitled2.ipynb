```python
#Importing necessary libraries
import scipy as sc
import numpy as np
import matplotlib.pyplot as plt

#State the initial condition
y0 = 1

#State the interval for t
start = 0
end = 2
t = np.linspace(start,end,end)

#State the differential equation
def dYdt(y,t):
    dydt = -2 * y * t
    return dydt

yinitial = np.linspace(y0,y0,end)

#Integration of differential equation
y = sc.integrate.odeint (dYdt,y0,t)

#Plotting the function
plt.figure()
plt.plot(t,y,'r','change in y value')
plt.plot(t,yinitial,'b--',linewidth =2)
plt.title(f'(y(t) versus t')
plt.xlabel('Time (t)')
plt.ylabel('y value')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.legend()
plt.show()
```
